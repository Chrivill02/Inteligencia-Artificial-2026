```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from io import StringIO
```

```python
url = 'https://es.wikipedia.org/wiki/Organizaci%C3%B3n_territorial_de_Guatemala'
head = 'User-Agent: Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
```

```python
response = requests.get(url, headers={'User-Agent': header})
```

```python
print(f' Status code {url}: {response.status_code}')
```

```python
response.text[:500]  # Muestra los primeros 500 caracteres del HTML
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
tables = soup.find('table', {'class': 'wikitable'})
tables
```

```python
tables = pd.read_html(StringIO(str(tables)))
tables
```

```python
tables[1]
```

```python
departamentos = tables[1]
```

```python
departamentos.columns = ['Departamento', 'Cabecera', 'Superficie', 'Poblacion']
departamentos
```

```python
region = soup.find_all('table')[2]
region = pd.read_html(StringIO(str(region)))[0]
region
```
