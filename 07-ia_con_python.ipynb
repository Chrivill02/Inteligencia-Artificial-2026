```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# importar el separador de muustras para emntrenar y probar
from sklearn.model_selection import train_test_split
# importar el clasificador:arbol de decisi√≥n
from sklearn.tree import DecisionTreeClassifier, plot_tree
# importar las metricas para medir la eficiencia del modelo
from sklearn import metrics
```

```python
# Cargar el dataset
data = sns.load_dataset('iris')
```

```python
# Separacion de datos en Entremiento-prueba
train, test = train_test_split(data, test_size=0.4, random_state=42, stratify=data['species'])
```

```python
x_train = train[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y_train = train['species']
```

```python
fn = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
cn = ['setosa', 'versicolor', 'virginica']
```

```python
x_train = train[fn]
y_train = train['species']
```

```python
#Test de pruebas siguiendo el mismo formato que el de entrenamiento
x_test = test[fn]
y_test = test['species']
```

```python
#Arbol de decision
mod_dt = DecisionTreeClassifier(max_depth=3, random_state=112)
mod_dt.fit(x_train, y_train)
prediccion = mod_dt.predict(x_test)
```

```python
mod_dt.feature_importances_
```

```python
#Visualizar las reglas de clasificaci√≥n
plt.figure(figsize=(10,8))
plot_tree(mod_dt, feature_names=fn, class_names=cn, filled=True);
```

```python
# Eficiencia del modelo
eficiencia = metrics.accuracy_score(y_test, prediccion)
print(f'Eficiencia del modelo: {eficiencia}')
```
