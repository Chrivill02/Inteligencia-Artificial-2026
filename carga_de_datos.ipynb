```python
contingencia = pd.crosstab(tabla['Edad'], tabla['Sexo'])
contingencia
```

```python
n = len(tabla)
intervalos = pd.cut(tabla['Edad'], bins=5)
intervalos
```

```python
frecuencias = pd.crosstab(index=intervalos, columns='f')
frecuencias
```

```python
frecuencias['fa'] = frecuencias['f'].cumsum()
frecuencias['fr'] = frecuencias['f'] / n
frecuencias['fra'] = frecuencias['fr'].cumsum ()
frecuencias['%'] = frecuencias['fr'] * 100
frecuencias['%a'] = frecuencias['fra'] * 100
frecuencias
```

```python
frecuencias.index.categories.mid
frecuencias['Xm'] = frecuencias.index.categories.mid
frecuencias['f.Xm'] = frecuencias['f'] * frecuencias['Xm']
frecuencias
```

```python
print(frecuencias['f.Xm'].sum() / n)
```
